```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.gaussian_process import GaussianProcess
import pandas_datareader as data
from mpl_toolkits import mplot3d
import seaborn as sns; sns.set() 
plt.style.use('fivethirtyeight')
```

```python
aapl=data.DataReader('aapl',start='2012',end='2017',data_source='google')
```

```python
volume=aapl['Volume']
```

```python
volume=volume/1000000
volume.head()
```

```python
mVolume=volume.resample('M').mean()

dates=pd.DatetimeIndex(start='2012',end='2017',freq="A")
dates
```

```python
fig,ax=plt.subplots(2,figsize=(14,8))
fig.subplots_adjust(hspace=0.2)
#mVolume.plot(dates,mVolume,ax=ax[1])

aapl['Close'].plot(ax=ax[0])
ax[0].set_xlabel("")

mVolume.plot(x=dates,kind='bar',ax=ax[1],color='green')
ax[1].xaxis.set_major_locator(plt.NullLocator())
ax[1].set_ylabel("Monthly Average Volume (Millions)",size=10)
ax[1].set_xlabel("")
```
